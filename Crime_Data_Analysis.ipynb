```python
#import matplotlib and seborn for visual analysis
import matplotlib.pyplot as plt
import seaborn as sns
```

# Determine trends in crime occurrence over time.

```python
# Convert 'Date_Rptd' column to datetime format
df['Date_Rptd'] = pd.to_datetime(df['Date_Rptd'], format='%d-%m-%Y')

# Extract month, day of the week from the 'Date_Rptd' column
df['Month'] = df['Date_Rptd'].dt.month
df['Day_of_Week'] = df['Date_Rptd'].dt.dayofweek

# Monthly trends
monthly_crime_counts = df.groupby('Month').size()
plt.figure(figsize=(10, 6))
monthly_crime_counts.plot(kind='bar', color='skyblue')
plt.title('Monthly Crime Trends')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.xticks(range(0, 12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
```

```python
# weekly trends
monthly_crime_counts = df.groupby('Day_of_Week').size()
plt.figure(figsize=(10, 6))
monthly_crime_counts.plot(kind='bar', color='skyblue')
plt.title('Weekly Crime Trends')
plt.xlabel('Day_of_Week')
plt.ylabel('Number of Crimes')
plt.xticks(range(0, 7), ['mon','Tue','wed','Thus','Fri','Sat','Sun'])
```

# Where are the geographical hotspots for reported crimes?

```python
# mapping hotspot on lat and lon
plt.figure(figsize=(10, 8))
plt.scatter(df['LON'], df['LAT'], color='red', alpha=0.6)
plt.title('Crime Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

As per data most of the crime occur between longitude (-118.30 to -118.20) and latitude (34.0 to 34.2)

# Where do most crimes occur based on the "Location" column?

```python
# Group data by location and count the occurrences
crime_by_location = df['Location'].value_counts()

# Select the top n locations with the highest crime occurrences
n = 10  
top_locations = crime_by_location.head(n)

# Plotting
plt.figure(figsize=(10, 6))
top_locations.plot(kind='bar', color='skyblue')
plt.title('Top {} Locations with Most Reported Crimes'.format(n))
plt.xlabel('Location')
plt.ylabel('Number of Reported Crimes')
plt.xticks(rotation=20, ha='right')
plt.tight_layout()
plt.show()
```

As per above data '800 N ALAMEDA ST', '700 W 7TH ST', '1100 S FIGUEROA ST' , '300 E 5TH ST', '6TH', '100 E 5TH ST', '600 S SPRING ST', '500 S SAN PEDRO ST', '7TH ST' AND '700 S FLOWER ST' this are the top 10 location where most of the crime occur.

# What is the distribution of victim ages in reported crimes?

```python
# Plotting the distribution of victim ages
plt.figure(figsize=(10, 6))
sns.histplot(df['Vict_Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Victim Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

As per above data - Most of the victim ages are between 0-10, 25-35 and nearly 55-60 age.

# Is there a significant difference in crime rates between male and female victims?


```python
# Plotting the distribution of victim genders
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Vict_Sex', palette='pastel')
plt.title('Distribution of Victim Genders')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

The graph illustrates a noticeable disparity in reported crime occurrences between male and female victims. With male victims depicted in a higher frequency compared to female victims, it suggests a significant disproportion in crime victimization by gender.

# Identify common premises descriptions where crimes occur.

```python
# Plotting the common premises descriptions
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Premis_Desc', order=df['Premis_Desc'].value_counts().index[:10], palette='viridis')
plt.title('Common Premises Descriptions Where Crimes Occur')
plt.xlabel('Count')
plt.ylabel('Premises Description')
plt.show()
```

# What is the distribution of reported crimes based on Crime Code?

```python
# Group data by crime code and count the occurrences
crime_by_code = df['Crm_Cd_Desc'].value_counts()

# Select the top n crime codes with the highest occurrences
n = 5  
top_crime_codes = crime_by_code.head(n)

# Plotting
plt.figure(figsize=(10, 6))
top_crime_codes.plot(kind='bar', color='skyblue')
plt.title('Top {} Reported Crimes by Crime Code'.format(n))
plt.xlabel('Crime Code Description')
plt.ylabel('Number of Reported Crimes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
